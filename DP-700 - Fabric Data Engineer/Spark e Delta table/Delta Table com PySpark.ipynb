## Estudos em PySpark

```python
df = spark.sql("SELECT A.NROEMPRESA, A.SEQLOCAL, A.LOCAL FROM HML_GL_02.RAW_MRL_LOCAL A LIMIT 10")
display(df)
```

## Criando Tabela

```python
df.write.mode("overwrite").saveAsTable("OBA_MRL_LOCAL")
```

## Criando tabela a partir de um select

```python
df = spark.sql("""
    SELECT A.NROEMPRESA, A.SEQLOCAL, A.LOCAL
    FROM HML_GL_02.RAW_MRL_LOCAL A
""")

df.write.format("delta").mode("overwrite").saveAsTable("OBA_MRL_LOCAL_DELTA")
```

### Drop Table

```python
spark.sql("DROP TABLE IF EXISTS OBA_MRL_LOCAL")
```

### Select usando PySpark

```python
df = spark.sql("""
    SELECT *
    FROM HML_GL_02.OBA_MRL_LOCAL_DELTA
""")
df.show()
```

```python
df = spark.table("HML_GL_02.OBA_MRL_LOCAL_DELTA").select("NROEMPRESA", "SEQLOCAL", "LOCAL")
df.show()
```
